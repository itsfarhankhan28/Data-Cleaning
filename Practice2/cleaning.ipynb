```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Windows 10\Desktop\Data_Preprocessing_Cleaning\Practice2\Practice_dataset1.xlsx")
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.fillna('')
df
```

```python
#convertiing to datetime format

datetime_format = "%M-%d-%y"

df['Join_Date'] = df['Join_Date'].apply(lambda x: pd.to_datetime(x, format=datetime_format))
df
```

```python
print(df['Salary'].dtypes)
```

```python
df['Salary'] = df['Salary'].apply(lambda x: pd.to_numeric(x))
df
```

```python
print(df['Salary'].dtypes)
```

```python
df
```

```python
filtered_salary = df[df['Salary'] > 50000]
filtered_salary
```

```python
filtered_gender = df[df['Gender'] == 'Female']
filtered_gender
```

```python
df
```

```python
df = df.fillna('Unknown')
df
```

```python
df['Name'] = df['Name'].apply(lambda x: 'Unknown' if x == '' else x)
df
```

```python
df['Email'] = df['Email'].apply(lambda x: 'Unknown' if x == '' else x)
df 
```

```python
import datetime

current_day = datetime.date.today()
print(current_day)
# year = day.year
# print(year)
```

```python
for i in df.index:
    if df.loc[i, 'Join_Date'] == 'Unknown':
        df.drop(i, inplace=True)

df
```

```python
df['Years_of_Service'] = df['Join_Date'].apply(lambda x: ((current_day - x.date()).days/365.25))

df['Years_of_Service'] = df['Years_of_Service'].round(2)
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.to_excel('./cleaned_dataset1.xlsx')
```
