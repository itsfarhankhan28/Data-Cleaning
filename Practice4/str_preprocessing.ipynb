```python
import pandas as pd
```

```python
df = pd.read_excel('./dataset.xlsx')
df
```

```python
df[['Email','Domain']] = df['Email'].str.split('@', n=1 ,expand=True)
df
```

```python
df[['First_Name','Last_Name']] = df['Name'].str.split(' ', n=1, expand=True)
df
```

```python
df['Clean_Phone'] = df['Phone_Number'].str.replace('-',' ')
df
```

```python
df["City"] = df["Address"].str.extract(r",\s*([^,]+),")
df
```

```python
df['Word_Count'] = df['Address'].str.split()
df['Word_Count']
```

```python
df['Domain'] = df['Domain'].str.replace('.','_')
df['Domain']
```

```python
# df['Name'] = df['Name'].str.upper()
# print(df['Name'].dtype)

df['Name']=df['Name'].astype('object')
```

```python
print(df['Name'].dtype)
```

```python
df['Name'] = df['Name'].str.upper()
df['Name']
```

```python
filtered_name = df[df['Name'].str.contains('LEE')]
filtered_name
```

```python
df['Full_Address'] = df[['Address','City']].agg(','.join, axis=1)
df['Full_Address']
```
