```python
grouped = df.groupby('Department')

for name,group in grouped:
    print("Name: ", name)
    print(group)
    print('\n')
```

```python
aggregate_avg = grouped.agg({
    'Salary':'mean'
})

print(aggregate_avg)
```

```python
aggregated_max = grouped.agg({
    'Salary':'sum'
})

print(aggregated_max)
```

```python
dapt = aggregated_max.idxmax()
print(dapt)
```

```python
dept_people = grouped.agg({
    'Department':'count'
})

print(dept_people)
```
