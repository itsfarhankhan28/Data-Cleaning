```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Windows 10\Desktop\Data_Preprocessing_Cleaning\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('./_')
df
```

```python
#Point to note in this latest version we need to add regular expression as True.

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True) 
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# new way to add number of columns i.e n=2

df[["Street Name","City","Pin Code"]] = df['Address'].str.split(',',n=2,expand=True)
df
```

```python
df = df.drop(columns = 'Address')
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
for i in df.index:
    if df.loc[i, "Do_Not_Contact"] == "Y":
        df.drop(i, inplace=True)

df
```

```python
df = df.fillna('')
```

```python
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
for i in df.index:
    if df.loc[i, 'Phone_Number'] == "nan--":
        df.drop(i, inplace=True)

df
```

```python
df = df.drop(columns = 'index')
df
```

```python
for i in df.index:
    if df.loc[i, 'Phone_Number'] == "Na--":
        df.drop(i, inplace=True)

df
```
