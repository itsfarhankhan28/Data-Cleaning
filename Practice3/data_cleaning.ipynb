```python
import pandas as pd
```

```python
df = pd.read_excel('./Customer_dataset.xlsx')
df
```

```python
df = df.fillna('Unknown')
df
```

```python
date_format = '%Y-%m-%d'

df['Order_Date'] = df['Order_Date'].apply(lambda x: pd.to_datetime(x, format=date_format))
df
```

```python
df['Price'] = df['Price'].apply(lambda x: pd.to_numeric(x))
df['Discount'] = df['Discount'].apply(lambda x: pd.to_numeric(x))
df
```

```python
df['Price'] = df['Price'].astype('int64')

# print(df['Price'].dtype)
df['Total_Amount'] = df[['Quantity','Price','Discount']].apply(lambda x: x['Quantity'] * x['Price'] - x['Discount'], axis=1)
df
```

```python
fileterd_status = df[df['Delivery_Status'] == 'Delivered']
fileterd_status
```

```python
fileterd_Price = df[df['Price'] > 100]
fileterd_Price
```

```python
grouped = df.groupby('Product_Category')

for name, group in grouped:
    print('name: '+ name)
    print(group)
    print('\n')
```

```python
aggregated = grouped.agg({
    'Total_Amount':'sum'
})

print(aggregated)
```
